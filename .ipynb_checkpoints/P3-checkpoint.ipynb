# Question 1 - is there any geographic pattern to the distribution of non-functional wells?

- If there's certain regions of Tanzania that may be in dire need of repairs.

```python
#build a map of the status of the wells 
train_set['Well State'] = train_set['status_group'].map(lambda x: 'Functional' if x.lower() == 'functional'
                                         else 'Non-Functional' if x.lower() =='non functional'
                                         else 'Needs Repair')
fig = plt.figure(figsize=(15,15))
sns.set_style('whitegrid')
graph_palette = {'Non-Functional':'r', 'Functional':'b', 'Needs Repair':"y"}
geo_status_chart = sns.scatterplot(data=train_set, x='longitude', y='latitude', hue='Well State', alpha=0.5, palette=graph_palette)
sns.despine(left=True, bottom=True)
plt.title('Well Status Throughout Tanzania', size='xx-large')
plt.xlabel('Longitude', size='x-large')
plt.ylabel('Latitude', size='x-large')
plt.yticks(ticks=geo_status_chart.get_yticks(), size='large')
plt.xticks(ticks=geo_status_chart.get_xticks(), size='large')
plt.ylim(bottom=-12, top=-0.2)
plt.xlim(left=29, right=41)
plt.show()
```

# Question 2 - Does the kind of extraction method effect the functionality of the well?

```python
#create function to easily compare the 'status group' between by percentage

def assess_categorical_correlation_dict(column_name, dataframe):
    working_total = []
    broken_total = []
    fixable_total = []

    cat_dicts = []

    for value in dataframe[column_name].unique():
        all_working = len(dataframe.loc[(dataframe[column_name] == value) & (dataframe['status_group'] == 'functional')])
        all_broken = len(dataframe.loc[(dataframe[column_name] == value) & (dataframe['status_group'] == 'non functional')])
        all_fixable = len(dataframe.loc[(dataframe[column_name] == value) & (dataframe['status_group'] == 'functional needs repair')])
        total_val = len(dataframe.loc[dataframe[column_name] == value])

        working_perc = all_working / total_val
        broken_perc = all_broken / total_val
        fixable_perc = all_fixable / total_val

        working_total.append(working_perc)
        broken_total.append(broken_perc)
        fixable_total.append(fixable_perc)

        single_cat_dict = {}
        single_cat_dict['value'] = value.title()
        single_cat_dict['Percent Working'] = working_perc
        single_cat_dict['Percent Non Functional'] = broken_perc
        single_cat_dict['Percent Needing Repair'] = fixable_perc
        cat_dicts.append(single_cat_dict) 
    
    totals_dict = {'value': 'overall', 'Percent Working': np.array(working_total).mean(), 'Percent Non Functional': np.array(broken_total).mean(),
                                        'Percent Needing Repair': np.array(fixable_total).mean()}
    cat_dicts.append(totals_dict)
    return cat_dicts

#graph to show comparision
def plot_comparison_chart(category, df, overall=False, norm_line=False, cmap='Pastel1'):
    comp_dict_list = assess_categorical_correlation_dict(category, df)
    if not overall:
        comp_dict_list = comp_dict_list[0:-1]
    comp_df = pd.DataFrame(comp_dict_list)
    comp_df = comp_df.set_index('value')
    comp_df = comp_df.sort_values('value')
    ax = comp_df.plot.bar(rot=0, figsize=(20,10), cmap=cmap)
    if norm_line:
        plt.axhline(label='Overall Well Functionality Rate', y=.5466, dash_joinstyle='miter', color='gray', ls='--')
    plt.legend()
    plt.ylabel('Percent', size='x-large')
    ax.tick_params(axis='x', labelsize=12)
    vals = ax.get_yticks()
    ax.set_yticklabels(['{:.0%}'.format(x) for x in vals], fontsize='large')
    return ax

#Extraction type
train_set['extraction_type_class'] = train_set['extraction_type_class'].map(lambda x: 'other' if x.lower() == 'wind-powered' 
                                                                                else 'other' if x.lower() == 'rope pump' 
                                                                                else x)
# call the graphing function
sns.set_style('darkgrid')
extraction_chart = plot_comparison_chart('extraction_type_class', train_set, overall=False, norm_line=True, cmap='tab10')
extraction_chart.set_title('Percent of Functioning Wells by Extraction Class', size='xx-large')
extraction_chart.set_xlabel('Extraction Class', size='x-large')
extraction_chart.set_ylim(top=0.79)
plt.show()
```

The most common used extraction methods work well. Gravity and Handpump are the most popular and appear to be functioning when checked. Motorpump is commonly used but appears to be non-functional when checked. Moving foward; gravity, handpump and maybe even submersible sould be used most often, depending on the situation.

Next we look at the waterpoint types to see if there is any similar connection there.

# Question 3 - Does waterpoint type effect well functionality?

```python
# how about waterpoint?
#consolidate the oddities together
train_set['waterpoint_type_group'] = train_set['waterpoint_type_group'].map(lambda x: 'dam, trough, or spring' if x.lower() =='dam'
                                                                         else 'dam, trough, or spring' if x == 'cattle trough'
                                                                         else 'dam, trough, or spring' if x == 'improved spring'
                                                                         else x)
waterpoint_chart = plot_comparison_chart('waterpoint_type_group', train_set, overall=False, norm_line=True, cmap='tab10')
waterpoint_chart.set_title("Percent of Functioning Well by Waterpoint", size='xx-large')
waterpoint_chart.set_xlabel('Waterpoint Type', size='x-large')
plt.show()
```

It just so happens the two most common water point types have the best functionalty. There are a few rarer ones like, dams and improved springs, that have high functionality. However, the resources for those are even more limited which is why there are so few.

This shows that there are certain water points that should be focused on in the future when possible.

# Question 5 - Does a well's age influence its functionality?
Logically: the older it is, the less water a well would have. Lets see.

```python
#drop all a 0's
construction_years_df = train_set.drop(train_set.loc[train_set['construction_year'] == 0].index)

# make a dictionary of years and their percentage of functioning wells
year_dict = {}
for year in construction_years_df['construction_year'].unique():
    year_dict[year] = len(construction_years_df.loc[(train_set['construction_year'] == year) & (construction_years_df['status_group'] ==      'functional')]) / len(construction_years_df.loc[train_set['construction_year'] == year])

# plot the results
fig = plt.figure(1, figsize=(20, 10))
sns.lineplot(x=list(year_dict.keys()), y=list(year_dict.values()), linewidth=3)
plt.title('Percent of Functional Wells by Year of Construction', size='xx-large')
plt.xlabel('Year', size='x-large')
plt.ylabel('Percentage', size='x-large')
plt.xticks(rotation=25, fontsize='large')
plt.yticks(fontsize='large')
plt.axhline(label='Overall Well Functionality Rate', y=.5466, dash_joinstyle='miter', color='gray', ls='--')
plt.legend()
plt.show()
```
